```python
# Unfortunately, unlike R, `waffle` charts are not built into any of the Python visualization libraries. Therefore, we will learn how to create them from scratch.
# Unfortunately, unlike R, `waffle` charts are not built into any of the Python visualization libraries. Therefore, we will learn how to create them from scratch.
# **Step 1.** The first step into creating a waffle chart is determing the proportion of each category with respect to the total.
```

```python
# compute the proportion of each category with respect to the total
total_values = df_dsn['Total'].sum()
category_proportions = df_dsn['Total'] / total_values

# print out proportions
pd.DataFrame({"Category Proportion": category_proportions})
```

```python
# Step 2. The second step is defining the overall size of the waffle chart.
width = 40 # width of chart
height = 10 # height of chart

total_num_tiles = width * height # total number of tiles

print(f'Total number of tiles is {total_num_tiles}.')
```

```python
#Step 3. The third step is using the proportion of each category to determe it respective number of tiles
# compute the number of tiles for each category
tiles_per_category = (category_proportions * total_num_tiles).round().astype(int)

# print out number of tiles per category
pd.DataFrame({"Number of tiles": tiles_per_category})
```

```python
#Based on the calculated proportions, Denmark will occupy 129 tiles of the waffle chart, Norway will occupy 77 tiles, and Sweden will occupy 194 tiles.
#Step 4. The fourth step is creating a matrix that resembles the waffle chart and populating it.
```

```python
# initialize the waffle chart as an empty matrix
waffle_chart = np.zeros((height, width), dtype = np.uint)

# define indices to loop through waffle chart
category_index = 0
tile_index = 0

# populate the waffle chart
for col in range(width):
    for row in range(height):
        tile_index += 1

        # if the number of tiles populated for the current category is equal to its corresponding allocated tiles...
        if tile_index > sum(tiles_per_category[0:category_index]):
            # ...proceed to the next category
            category_index += 1

        # set the class value to an integer, which increases with class
        waffle_chart[row, col] = category_index

print ('Waffle chart populated!')
```

```python
# Now it would very inefficient to repeat these seven steps every time we wish to create a waffle chart. So let's combine all seven steps into one function called create_waffle_chart. This function would take the following parameters as input:
# categories: Unique categories or classes in dataframe.
# values: Values corresponding to categories or classes.
# height: Defined height of waffle chart.
# width: Defined width of waffle chart.
# colormap: Colormap class
# value_sign: In order to make our function more generalizable, we will add this parameter to address signs that could be associated with a value such as %, $, and so on. value_sign has a default value of empty string
```

```python
def create_waffle_chart(categories, values, height, width, colormap, value_sign=''):

    # compute the proportion of each category with respect to the total
    total_values = sum(values)
    category_proportions = [(float(value) / total_values) for value in values]

    # compute the total number of tiles
    total_num_tiles = width * height # total number of tiles
    print ('Total number of tiles is', total_num_tiles)

    # compute the number of tiles for each catagory
    tiles_per_category = [round(proportion * total_num_tiles) for proportion in category_proportions]

    # print out number of tiles per category
    for i, tiles in enumerate(tiles_per_category):
        print (df_dsn.index.values[i] + ': ' + str(tiles))

    # initialize the waffle chart as an empty matrix
    waffle_chart = np.zeros((height, width))

    # define indices to loop through waffle chart
    category_index = 0
    tile_index = 0

    # populate the waffle chart
    for col in range(width):
        for row in range(height):
            tile_index += 1

            # if the number of tiles populated for the current category
            # is equal to its corresponding allocated tiles...
            if tile_index > sum(tiles_per_category[0:category_index]):
                # ...proceed to the next category
                category_index += 1

            # set the class value to an integer, which increases with class
            waffle_chart[row, col] = category_index

    # instantiate a new figure object
    fig = plt.figure()

    # use matshow to display the waffle chart
    colormap = plt.cm.coolwarm
    plt.matshow(waffle_chart, cmap=colormap)
    plt.colorbar()

    # get the axis
    ax = plt.gca()

    # set minor ticks
    ax.set_xticks(np.arange(-.5, (width), 1), minor=True)
    ax.set_yticks(np.arange(-.5, (height), 1), minor=True)

    # add dridlines based on minor ticks
    ax.grid(which='minor', color='w', linestyle='-', linewidth=2)

    plt.xticks([])
    plt.yticks([])

    # compute cumulative sum of individual categories to match color schemes between chart and legend
    values_cumsum = np.cumsum(values)
    total_values = values_cumsum[len(values_cumsum) - 1]

    # create legend
    legend_handles = []
    for i, category in enumerate(categories):
        if value_sign == '%':
            label_str = category + ' (' + str(values[i]) + value_sign + ')'
        else:
            label_str = category + ' (' + value_sign + str(values[i]) + ')'

        color_val = colormap(float(values_cumsum[i])/total_values)
        legend_handles.append(mpatches.Patch(color=color_val, label=label_str))

    # add legend to chart
    plt.legend(
        handles=legend_handles,
        loc='lower center',
        ncol=len(categories),
        bbox_to_anchor=(0., -0.2, 0.95, .1)
    )
    plt.show()
```

```python
def create_waffle_chart(categories, values, height, width, colormap, value_sign=''):
    """
    Creates a waffle chart visualization

    Parameters:
    -----------
    categories: Unique categories or classes in dataframe
    values: Values corresponding to categories
    height: Height of waffle chart
    width: Width of waffle chart
    colormap: Matplotlib colormap
    value_sign: Value sign (e.g., %, $) - default ''
    """

    import numpy as np
    import matplotlib.pyplot as plt
    import matplotlib.patches as mpatches

    # Compute the proportion of each category
    total_values = sum(values)
    category_proportions = [(float(value)/total_values) for value in values]

    # Compute total number of tiles
    total_tiles = width * height
    tiles_per_category = [round(proportion * total_tiles) for proportion in category_proportions]

    # Initialize waffle chart
    waffle_chart = np.zeros((height, width), dtype=np.uint)

    # Populate waffle chart
    category_index = 0
    tile_index = 0

    for col in range(width):
        for row in range(height):
            tile_index += 1

            if tile_index > sum(tiles_per_category[0:category_index]):
                category_index += 1

            waffle_chart[row, col] = category_index

    # Create figure
    fig = plt.figure()
    plt.matshow(waffle_chart, cmap=colormap)
    plt.colorbar()

    # Formatting
    ax = plt.gca()
    ax.set_xticks(np.arange(-.5, width, 1), minor=True)
    ax.set_yticks(np.arange(-.5, height, 1), minor=True)
    ax.grid(which='minor', color='w', linestyle='-', linewidth=2)
    plt.xticks([])
    plt.yticks([])

    # Compute cumulative sum and create legend
    values_cumsum = np.cumsum(values)
    total_values = values_cumsum.iloc[-1] if hasattr(values_cumsum, 'iloc') else values_cumsum[-1]

    legend_handles = []
    for i, category in enumerate(categories):
        label_str = f"{category} ({value_sign}{values.iloc[i] if hasattr(values, 'iloc') else values[i]})"
        color_val = colormap(float(values_cumsum.iloc[i] if hasattr(values_cumsum, 'iloc') else values_cumsum[i])/total_values)
        legend_handles.append(mpatches.Patch(color=color_val, label=label_str))

    plt.legend(
        handles=legend_handles,
        loc='lower center',
        ncol=len(categories),
        bbox_to_anchor=(0., -0.2, 0.95, .1)
    )

    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# First, ensure you have the correct data
# Let's create sample data if df_can isn't available
data = {'Country': ['China', 'India'], 'Total': [50000, 75000]}  # Example numbers
df_ci = pd.DataFrame(data).set_index('Country')

def create_waffle_chart(categories, values, height, width, colormap, value_sign=''):
    # Calculate proportions
    total = sum(values)
    proportions = [float(v)/total for v in values]

    # Calculate number of tiles per category
    total_tiles = width * height
    tiles_per_category = [round(p * total_tiles) for p in proportions]

    # Create empty waffle chart
    waffle = np.zeros((height, width), dtype=np.uint)

    # Populate waffle chart
    cat_index = 0
    tile_index = 0

    for col in range(width):
        for row in range(height):
            tile_index += 1
            if tile_index > sum(tiles_per_category[:cat_index]):
                cat_index += 1
            waffle[row, col] = cat_index

    # Create figure and axis explicitly
    fig, ax = plt.subplots(figsize=(10, 6))

    # Create visualization
    ax.matshow(waffle, cmap=colormap)
    plt.colorbar(ax.matshow(waffle, cmap=colormap), shrink=0.4)

    # Formatting
    ax.set_xticks(np.arange(-.5, width, 1), minor=True)
    ax.set_yticks(np.arange(-.5, height, 1), minor=True)
    ax.grid(which='minor', color='w', linestyle='-', linewidth=2)
    ax.set_xticks([])
    ax.set_yticks([])

    # Create legend
    legend_handles = []
    for i, cat in enumerate(categories):
        label = f"{cat} ({value_sign}{values[i]})"
        color = colormap(float(i)/len(categories))
        legend_handles.append(mpatches.Patch(color=color, label=label))

    ax.legend(handles=legend_handles,
              loc='lower center',
              bbox_to_anchor=(0.5, -0.2),
              ncol=len(categories))

    ax.set_title('Proportion of Immigrants from China and India (1980-2013)', pad=20)
    plt.tight_layout()
    plt.show()

# Create the chart
create_waffle_chart(categories=df_ci.index.tolist(),
                    values=df_ci['Total'].tolist(),
                    height=8,
                    width=12,
                    colormap=plt.cm.Accent,
                    value_sign='')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (in a real scenario, you'd load your actual dataset)
# Let's create a DataFrame with sample immigration data
years = list(range(1980, 2014))
data = {
    'Year': years * 3,
    'Country': ['Denmark']*len(years) + ['Sweden']*len(years) + ['Norway']*len(years),
    'Immigrants':
        # Denmark - synthetic data
        [int(100 + x*2 + x**2*0.03 - x**3*0.0002) for x in range(len(years))] +
        # Sweden - synthetic data
        [int(150 + x*3 + x**2*0.04 - x**3*0.0003) for x in range(len(years))] +
        # Norway - synthetic data
        [int(80 + x*1.5 + x**2*0.02 - x**3*0.0001) for x in range(len(years))]
}

df = pd.DataFrame(data)

# Create the scatter plot with regression lines
plt.figure(figsize=(12, 6))
sns.set_style('whitegrid')

# Scatter plot with regression lines, hue by country
scatter = sns.lmplot(
    x='Year',
    y='Immigrants',
    data=df,
    hue='Country',
    palette=['#C8102E', '#004B87', '#EF2B2D'],  # Colors matching country flags
    height=6,
    aspect=1.5,
    scatter_kws={'s': 50, 'alpha': 0.7},
    line_kws={'lw': 2}
)

# Customize the plot
plt.title('Immigration from Scandinavia to Canada (1980-2013)', fontsize=14, pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Immigrants', fontsize=12)
plt.grid(True, alpha=0.3)

# Adjust x-axis ticks to show every 5 years
plt.xticks(range(1980, 2014, 5))

plt.tight_layout()
plt.show()
```
